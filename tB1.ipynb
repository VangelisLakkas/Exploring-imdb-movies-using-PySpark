Runtime variable has the abbrevation "min" which stands for minutes next to the number which declares the duration of the movie. To make this variable 
a useful input for my model I built a function which removes the commas and the "min" expression and replaces them with a space. After that I setted the data type to float

```python
def clean_runtime(value):
    if value is not None:
        value = value.replace(",", "")
        return float(value.replace("min", ""))
    else:
        return None
```

```python
clean_runtime_udf = F.udf(clean_runtime, FloatType())
```

```python
imdb_df = imdb_df.withColumn("runtime_cleaned", clean_runtime_udf("runtime"))
```

There are several null values in the runtime_cleaned variable. I found the median movie runtime and filled the null observations with this value

```python
pandas_imdb_clean['runtime_cleaned'].isnull().sum()
```

```python
median_runtime = imdb_df.selectExpr("percentile_approx(runtime_cleaned, 0.5)").collect()[0][0]
```

```python
imdb_df = imdb_df.withColumn("runtime_cleaned", F.when(col("runtime_cleaned").isNull(), median_runtime).otherwise(col("runtime_cleaned")))
```

I created a subset which contains only the metascore and its predictors

```python
predictors = ["metascore", "users_rating", "first_genre", "languages_count", "runtime_cleaned"]
subset_df = imdb_df.select(predictors)
```

```python
from pyspark.sql.functions import when
imdb_df = imdb_df.withColumn("first_genre", when(imdb_df["first_genre"] == "", "unknown").otherwise(imdb_df["first_genre"]))
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import VectorAssembler, OneHotEncoder, StringIndexer
from pyspark.ml import Pipeline
```

Since genre is a string I must encode it in order to create a categorical variable with multiple levels in my model

```python
indexer = StringIndexer(inputCol="first_genre", outputCol="first_genre_indexed")
```

```python
encoder = OneHotEncoder(inputCols=["first_genre_indexed"], outputCols=["first_genre_encoded"])
```

```python
assembler = VectorAssembler(inputCols=["users_rating", "languages_count", "first_genre_encoded", 'runtime_cleaned'], outputCol="features")
```

Creating and Fitting a pipeline which contains the genre encoding and creation of the matrix with the values of the predictors of the metascore

```python
pipeline = Pipeline(stages=[indexer, encoder, assembler])
```

```python
data = pipeline.fit(subset_df).transform(subset_df)
```

```python
data = data.select("features", "metascore")
```

Splitting the data into training and testing datasets

```python
train_data, test_data = data.randomSplit([0.75, 0.25], seed=123)
```

Defining the regression model and fitting into the training dataset

```python
regression = LinearRegression(featuresCol="features", labelCol="metascore")
```

```python
model = regression.fit(train_data)
```

Fitting the model into the testing dataset

```python
predictions = model.transform(test_data)
```

```python
actual_predicted = predictions.select("metascore", "prediction")
```

Calculating the r-squared statistic for evaluating the predictions that were made from the linear regression model

```python
evaluator = RegressionEvaluator(labelCol="metascore", predictionCol="prediction", metricName="r2")
r_squared = evaluator.evaluate(predictions)
```

R-squared is 45% which means that the model is mediocre and it is not the best statistical or machine learning model that can be defined

```python
print("R-squared:", r_squared)
```
